### **Categorical Data**

```python
plt.figure(1, figsize=(12, 8))
sns.countplot(data=df, x=df['Survived'])
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
df['Survived'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
sns.countplot(data=df, x='Survived')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
sns.countplot(data=df, x='Pclass')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
sns.countplot(data=df, x='Embarked')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f', shadow=True)
plt.show()
```

## **Numerical Data**

```python
plt.figure(1, figsize=(16, 10))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
sns.histplot(data=df, x='Age', kde=True)
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
sns.displot(data=df, x='Age', height=12, kde=True)
plt.show()
```

```python
plt.figure(1, figsize=(12, 8))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, y='Age', showmeans=True)
plt.subplot(1, 2, 2)
sns.boxplot(data=df, y='Fare', color='red', showmeans=True, width=0.5, linewidth=7)
plt.show()
```
