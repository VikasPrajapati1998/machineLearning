```python
def rename(name):
    if name=='Graduate':
        return 'Job Less'
    else:
        return name
```

### **Converters(columnName, funcName)**

```python
df = pd.read_csv("../data/train/aug_train.csv", sep=',', low_memory=False, converters={'education_level': rename})
df
```

### **Loading a huge dataset in chunks**

```python
dfs = pd.read_csv("../data/train/aug_train.csv", sep=',', low_memory=False, chunksize=5000)
for k in dfs:
    print(k)
    print("----------------------------------------------------------------")
```
