```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df
```

### **1. Scatterplot(Numerical- Numerical)**

```python
plt.figure(1, figsize=(12, 8))
sns.scatterplot(data=df, x='total_bill', y='tip', color='r')
plt.figure(2, figsize=(12, 8))
sns.boxplot(data=df, y='total_bill', showmeans=True)
plt.figure(2, figsize=(12, 8))
plt.figure(3, figsize=(12, 8))
sns.violinplot(data=df, y='total_bill', color='g', inner='box', width=0.5, showmeans=True)
plt.show()
```

```python
plt.figure(1, figsize=(16, 10), facecolor='aqua')
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', size='sex', sizes=(200, 200))
plt.show()

plt.figure(2, figsize=(16, 10), facecolor='teal')
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='smoker', size='sex', sizes=(200, 200))
plt.show()

plt.figure(3, figsize=(16, 10), facecolor='lime')
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex', style='smoker', size='size', sizes=(50, 200))
plt.show()
```

```python
df = pd.read_csv("../data/titanic/titanic.csv")
df
```

```python
df.sample(5)
```

```python
plt.figure(1, figsize=(12, 8), facecolor='olive')
sns.barplot(data=df, x='Pclass', y='Age')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8), facecolor='aqua')
sns.barplot(data=df, x='Pclass', y='Fare')
plt.show()
```

```python
plt.figure(1, figsize=(12, 8), facecolor='olive')
sns.barplot(data=df, x='Pclass', y='Fare', hue='Sex')
plt.show()
```

### **3. Box Plot(Numerical-Categorical)**

```python
plt.figure(1, figsize=(12, 8), facecolor='pink')
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived')
plt.show()

plt.figure(2, figsize=(12, 8), facecolor='aqua')
sns.boxplot(data=df, x='Sex', y='Age', hue='Survived', notch=True, palette='cool')
plt.show()
```

### **4. Distplot(Numerical-Categorical)**

```python
plt.figure(1, figsize=(12, 8), facecolor='skyblue')
sns.displot(data=df, x='Age', hue='Survived', height=12)
plt.show()

plt.figure(2, figsize=(12, 8), facecolor='skyblue')
sns.displot(data=df, x='Age', hue='Survived', height=12, kde=True)
plt.show()


plt.figure(3, figsize=(12, 8), facecolor='skyblue')
sns.displot(data=df, x='Age', hue='Survived', height=12, kind='kde')
plt.show()
```

```python
plt.figure(3, figsize=(12, 8), facecolor='skyblue')
sns.displot(data=df, x='Age', hue='Survived', height=12, rug=True, kde=True)
plt.show()
```

### **5. Heatmap(Categorical-Categorical)**

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
plt.figure(3, figsize=(12, 8), facecolor='skyblue')
sns.heatmap(pd.crosstab(df['Pclass'], df['Survived']), annot=True)
plt.show()
```

```python
df.groupby('Pclass').mean(numeric_only=True)['Survived']*100
```

```python
(df.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
plt.show()
```

```python
pd.crosstab(df['Embarked'], df['Survived'])
```
