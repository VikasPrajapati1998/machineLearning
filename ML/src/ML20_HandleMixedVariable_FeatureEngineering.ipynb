## **pandas.to_numeric**
- `pandas.to_numeric(arg, errors='raise', downcast=None, dtype_backend=_NoDefault.no_default)`
  - Convert argument to a numeric type.
- __arg__ : _scalar_, _list_, _tuple_, _1D array_, or _Series_
  - Argument to be converted.
- __errors__ : _ignore_, _raise_, _coerce_, default=_raise_
  - If â€˜raiseâ€™, then invalid parsing will raise an exception.
  - If â€˜coerceâ€™, then invalid parsing will be set as NaN.
  - If â€˜ignoreâ€™, then invalid parsing will return the input.
- __downcast__ : _str_, default = None
  - Can be 'integer', 'signed', 'unsigned', or 'float'. If not None, and if the data has been successfully cast to a numerical dtype (or if the data was numeric to begin with), downcast that resulting data to the smallest numerical dtype possible according to the following rules :
  - 'integer' or 'signed' : smallest signed int dtype (min : np.int8)
  - 'unsigned' : smallest unsigned int dtype (min : np.unit8)
  - 'float' : smallest float dtype (min : np.float32)
- __dtype_backend__ : _numpy_nullable_, _pyarrow_, defaults to NumPy backed DataFrames

```python
# Extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)
df.head()
```

### **Handling the Cabin Column**

```python
df.Cabin.unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['cabin_cat'] = df['Cabin'].str[0] # capture the first letter

df.head()
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors='coerce', downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])

df.head(20)
```
